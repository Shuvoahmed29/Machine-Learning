```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("Home Price.csv")
df
```

```python
plt.scatter(df.area,df.price,color='red',marker='+')
plt.xlabel("Area(Sqrt Ft)")
plt.ylabel("Price(US$)")
```

```python
area_df = df.drop('price',axis='columns')
area_df
```

```python
price_df = df.price
price_df
```

```python
reg = linear_model.LinearRegression()
reg.fit(area_df,price_df)
```

Predict price of a home with area = 3300 sqr ft

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
