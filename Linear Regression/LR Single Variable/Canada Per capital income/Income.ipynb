```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
plt.scatter(df.year,df.per_capital_income,color='red',marker='+')
plt.xlabel("Year")
plt.ylabel("Per Capital Income(US$)")
```

```python
year_df = df.drop('per_capital_income',axis='columns')
year_df
```

```python
income = df.per_capital_income
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(year_df,income)
```

```python
reg.predict([[2020]])
```
